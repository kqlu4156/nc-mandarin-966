```python
import random
import pandas as pd
import pickle
```

```python
# active/passive
exp_ap = [['content_id', 'sentence_id', 'sentence', 'plausible', 'rev', 'act']] 

sov = [
    ['热狗', '男孩', '吃'],
['木板', '女孩', '踢'],
['鸡蛋', '妈妈', '煮熟'],
['小孩', '爸爸', '举高'],
]

act = ['把','被','让']
count = 0

for i in range(5):
  elem = sov[i]
  item_dict = {'orig':{}, 'rev':{}}
  for a in act:
    s, o, v = elem
    if a == '把':
      exp_ap.append([i, count, f'{s}{a}{o}{v}了。', 0, 0, a])
      count += 1
      exp_ap.append([i, count, f'{o}{a}{s}{v}了。', 1, 1, a])
      count += 1
    else:
      exp_ap.append([i, count, f'{s}{a}{o}{v}了。', 1, 0, a])
      count += 1
      exp_ap.append([i, count, f'{o}{a}{s}{v}了。', 0, 1, a])
      count += 1

exp_ap = pd.DataFrame(exp_ap[1:],columns=exp_ap[0])
print(exp_ap)
exp_ap.to_csv('active_passive.csv')

svl = [
    ['小白', '生于', '北京'],
['王陵', '生于', '2002年'],
['员工', '就职于', '科技公司'],
['男孩', '就读于', '高校']
]

# time/loc
exp_tl = [['content_id', 'sentence_id', 'sentence', 'plausible']]

for i in range(4):
  elem = svl[i]
  s, v, l = elem
  exp_tl.append([i+5, count, f'{s}{v}{l}。', 1])
  count += 1
  exp_tl.append([i+5, count, f'{l}{v}{s}。', 0])
  count += 1

exp_tl = pd.DataFrame(exp_tl[1:],columns=exp_tl[0])
print(exp_tl)
exp_tl.to_csv('time_location.csv')

# past/future
exp_pf = [['content_id', 'sentence_id', 'sentence', 'plausible', 'time']]
time = ['昨天', '明天']
svo = [
    ['我','去', '学校'],
    ['她','去', '公园'],
    ['女孩', '吃', '面条']
]

for i in range(3):
  item_dict = {}
  elem = svo[i]
  s, v, o = elem
  for t in time:
    if t == '昨天':
      exp_pf.append([i+9, count, f'{t}{s}{v}了{o}。', 1, '昨天'])
      count += 1
      exp_pf.append([i+9, count, f'{t}{s}会{v}{o}。', 0, '昨天'])
      count += 1
    else:
      exp_pf.append([i+9, count, f'{t}{s}{v}了{o}。', 0, '明天'])
      count += 1
      exp_pf.append([i+9, count, f'{t}{s}会{v}{o}。', 1, '明天'])
      count += 1

exp_pf = pd.DataFrame(exp_pf[1:],columns=exp_pf[0])
print(exp_pf)
exp_pf.to_csv('past_future.csv')
```

```python
def select_item(id, plausible):
  if id < 5:
    return exp_ap.sample()
  elif id < 9:
    return exp_tl.sample()
  else:
    return exp_pf.sample()

def get_other_plausible_sents(cat_order):
  list_ap = list(exp_ap[~exp_ap['content_id'].isin(cat_order)]['sentence'])
  list_tl = list(exp_tl[~exp_tl['content_id'].isin(cat_order)]['sentence'])
  list_pf = list(exp_pf[~exp_pf['content_id'].isin(cat_order)]['sentence'])
  return list_ap + list_tl + list_pf
```

```python
def invert_item(id, item_dict):
  if random.getrandbits(1):
    # 50% chance displayed as-is
    return item_dict['sentence'].item()

  if id < 5:
    invert = exp_ap[exp_ap['content_id'] == int(item_dict['content_id'])]
    invert = invert[invert['plausible'] != int(item_dict['plausible'])]
    selection = invert.sample()
    return selection['sentence'].item()
  
  elif id < 9:
    invert = exp_tl[exp_tl['content_id'] == int(item_dict['content_id'])]
    invert = invert[invert['plausible'] != int(item_dict['plausible'])]
    selection = invert.sample()
    return selection['sentence'].item()
  
  else:
    invert = exp_pf[exp_pf['content_id'] == int(item_dict['content_id'])]
    invert = invert[invert['plausible'] != int(item_dict['plausible'])]
    selection = invert.sample()
    return selection['sentence'].item()

def select_parallel(id, item_dict):
  if id < 5:
    parallel = exp_ap[exp_ap['content_id'] == int(item_dict['content_id'])]
    parallel = parallel[parallel['sentence_id'] != int(item_dict['sentence_id'])]
    parallel = parallel[parallel['plausible'] != int(item_dict['plausible'])]
    selection = parallel.sample()
    return selection['sentence'].item()
  return None
```

```python
def generate_experiment():
  num_rounds = 6
  num_exp_items = 12
  num_exp_per_round = 2
  num_filler_per_round = 4
  num_implausible_exp = 6
  num_q_per_round = 10

  experiment_cat_order = list(range(num_exp_items))
  random.shuffle(experiment_cat_order)
  experiment_plausibility = [0]*num_implausible_exp + [1]*(num_exp_items - num_implausible_exp)
  random.shuffle(experiment_plausibility)
  experiment_items = []
  experiment_sents = []
  
  for i in range(num_exp_items):
    experiment_items.append(select_item(experiment_cat_order[i], experiment_plausibility[i]))
    experiment_sents.append(experiment_items[i]['sentence'].item())
  
  #print(experiment_items)

  filler_audio_sents = filler_items.copy()
  random.shuffle(filler_audio_sents)
  filler_audio_sents = filler_audio_sents[:num_rounds*num_filler_per_round]

  audio_list = []
  selection_list = []

  for i in range(num_rounds):
    print(f'ROUND {i+1}')
    exp_slice = slice(num_exp_per_round*i, num_exp_per_round*(i+1))
    filler_slice = slice(num_filler_per_round*i, num_filler_per_round*(i+1))
    items = experiment_items[exp_slice]
    sents = experiment_sents[exp_slice]
    audio_fillers = filler_audio_sents[filler_slice]
    cat_order = experiment_cat_order[exp_slice]
    plausibility = experiment_plausibility[exp_slice]

    audio_sents = sents + audio_fillers
    # random.shuffle(audio_items)

    distractor_sents = [sent for sent in filler_items if sent not in audio_sents] +\
                      get_other_plausible_sents(cat_order)


    # for j in range(num_exp_items):
    #   if j not in cat_order:
    #     distractor_sents.extend(exp_plausible_sents[j])
    random.shuffle(distractor_sents)

    audio_display_sents = [invert_item(cat_order[x],items[x]) for x in range(num_exp_per_round)] +\
     [select_parallel(cat_order[x],items[x]) for x in range(num_exp_per_round)] + audio_fillers
    
    audio_display_sents = list(filter(lambda item: item is not None, audio_display_sents))
    random.shuffle(audio_display_sents)
    #print(len(audio_display_sents))
    #print(num_q_per_round)
    #print(distractor_sents)

    distractor_sents = distractor_sents[:num_q_per_round-len(audio_display_sents)]
    #print(distractor_sents)

    display_sents = audio_display_sents + distractor_sents
    random.shuffle(display_sents)

    print('in audio: ')
    for sent in audio_sents:
      print(sent)
    print('selection options: ')
    for sent in display_sents:
      print(sent)
    
    audio_list.append(audio_sents)
    selection_list.append(display_sents)
  
  return audio_list, selection_list

```

```python
for i in range(20):
  audio_list, selection_list = generate_experiment()
  with open(f'/content/audio_{str(i+1).zfill(2)}.pckl','wb') as outfile:
    pickle.dump(audio_list, outfile)
  with open(f'/content/selection_{str(i+1).zfill(2)}.pckl','wb') as outfile:
    pickle.dump(selection_list, outfile)
```
